# Create Pickle File (meta)

```python
import pandas as pd
import pickle
import numpy as np
```

```python
data = pd.read_csv("depression_sem_acento_2017.txt")
```

```python
data.head()
```

```python
users = list(data.user.unique())
```

```python
# list usuarios para dict
cmap = {}
for x in range(0,len(users)):
    cmap[x] = users[x]
```

```python
data["subs"] = "depression"
```

```python
subs = list(data.subs.unique())
```

```python
bmap = {}
```

```python
# list usuarios para dict
for x in range(0,len(subs)):
    bmap[x] = subs[x]
```

```python
bmap
```

```python
dataVocab = pd.read_csv("vocab.txt", sep = " ", header=None)
```

```python
del dataVocab[1]
```

```python
# remove elementos NAN
dataVocab = dataVocab[pd.notnull(dataVocab[0])]
```

```python
len(dataVocab)
```

```python
listVocabulario = list(dataVocab[0])
```

```python
wmap = {}
# list usuarios para dict
for x in range(0,len(listVocabulario)):
    wmap[x] = listVocabulario[x]
```

```python
len(wmap)
```

```python
wmap = {v: k for k, v in wmap.items()}
```

```python
len(wmap)
```

## Gravar arquivo

```python
import pickle
```

```python
listDict = [wmap, cmap, bmap]
```

```python
f = open('metadata.pkl', 'wb')
pickle.dump(listDict, f, protocol=2)
f.close()
```
